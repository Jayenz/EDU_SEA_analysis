###Objective
Showed the relationship of education-related data in Kaggle datasets, which is education data from countries in South East Asia

###Load Dataset & Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
#import data from Kaggle (URL: https://www.kaggle.com/datasets/nelgiriyewithana/world-educational-data)
df = pd.read_csv('Global_Education.csv',encoding='latin-1')
#Select country in South East Asia
head_columns = ['Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam']
SEA_df = df[df['Countries and areas'].isin(head_columns)]
SEA_df
```

###*Data Visualization*
Focus on the data that is most related to the unemployment rate.


#Analyzed Completition rates in each country.

```python
#Compared Completion rates in each country.
#Correlation of each completion column are not much difference in value so I select all completion to visuallization.

completion_rate_columns = ['Completion_Rate_Primary_Male', 'Completion_Rate_Primary_Female',
                            'Completion_Rate_Lower_Secondary_Male', 'Completion_Rate_Lower_Secondary_Female',
                            'Completion_Rate_Upper_Secondary_Male', 'Completion_Rate_Upper_Secondary_Female']

# Melt the 'Completion_Rate' column
melted_df = SEA_df.melt(id_vars=['Countries and areas'], value_vars=completion_rate_columns,
                        var_name='Completion_Rate_Edu_Gender', value_name='Completion Rate')

sns.lineplot(x='Countries and areas', y='Completion Rate', hue='Completion_Rate_Edu_Gender', data=melted_df)

plt.xlabel('Countries and areas')
plt.xticks(rotation=90)
plt.ylabel('Completion Rate')
plt.title('Plot of Completion Rates & Country and areas')

plt.show()
```

#Analyzed gender-based differences in education

```python
#Analyze gender-based differences in education
#Choose the out off school rate column with the highest correlation with the unemployment rate.

sns.set_style("whitegrid")
OOSR_Lower_Secondary_Age_columns = ['OOSR_Lower_Secondary_Age_Male', 'OOSR_Lower_Secondary_Age_Female']

melted_SEA_df_OOSR = SEA_df.melt(id_vars=['Countries and areas'], value_vars=OOSR_Lower_Secondary_Age_columns,
                        var_name='Gender', value_name='OOSR_Lower_Secondary_Age')

# bar plot
sns.barplot(melted_SEA_df_OOSR, x='Countries and areas', y='OOSR_Lower_Secondary_Age', hue='Gender')

#plot
plt.title('correlation between Out-of-School Rates and Country (OOSR_Lower_Secondary_Age)')
plt.xlabel('Countries and areas')
plt.ylabel('OOSR_Lower_Secondary_Age')
plt.xticks(rotation=90)
plt.legend()

plt.show()
```

#Analyzed proficiency and unemployed rate.

```python
#Proficiency analysis
#compared Proficiency_Reading in each country

Proficiency_Reading_columns = ['Grade_2_3_Proficiency_Reading', 'Primary_End_Proficiency_Reading','Lower_Secondary_End_Proficiency_Reading']

# Melt the Proficiency_columns
melted_df = SEA_df.melt(id_vars=['Countries and areas'], value_vars=Proficiency_Reading_columns,
                        var_name='Proficiency_Reading', value_name='Proficiency')
#scatter plot
sns.scatterplot(x='Countries and areas', y='Proficiency', hue='Proficiency_Reading', data=melted_df)
#plot
plt.xlabel('Countries and areas')
plt.xticks(rotation=90)
plt.ylabel('Proficiency_Reading')
plt.title('Plot of Completion Rates & Country and areas')

plt.show()
```

Some values do not appear in the graph. It means that the value is zero.

```python
#Proficiency analysis
#compared Proficiency_math in each country

Proficiency_Math_columns = ['Grade_2_3_Proficiency_Math', 'Primary_End_Proficiency_Math', 'Lower_Secondary_End_Proficiency_Math']

# Melt the Proficiency_columns
melted_df = SEA_df.melt(id_vars=['Countries and areas'], value_vars=Proficiency_Math_columns,
                        var_name='Proficiency_math', value_name='Proficiency')
#scatter plot
sns.scatterplot(x='Countries and areas', y='Proficiency', hue='Proficiency_math', data=melted_df)
#plot
plt.xlabel('Countries and areas')
plt.xticks(rotation=90)
plt.ylabel('Proficiency_math')
plt.title('Plot of Completion Rates & Country and areas')

plt.show()
```
